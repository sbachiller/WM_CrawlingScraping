```
maxDocuments = 500
originURL = "https://store.steampowered.com/games/"
frontierQueue = [originURL]

br = mechanize.Browser()
# br.set_handle_robots(False)
# br.addheaders = [("User-agent", "Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.9.2.13) Gecko/20101206 Ubuntu/10.10 (maverick) Firefox/3.6.13")]  
```

```
def generateRelativePath(link):
  newURL = "http://" + urlparse(link.base_url).netloc + link.url
  if (validators.url(newURL) == True and link.url[0] == '/'): 
    return newURL
```

```
def frontierQueueInsert(frontierQueue, url):
  toInsert = True

  if (url in frontierQueue):
    toInsert = False
  if (len(frontierQueue) >= maxDocuments):
    toInsert = False

  if (toInsert == True):
    frontierQueue.append(url)
  return frontierQueue
```

```
def crawl_url(br, url, frontierQueue):
  print("Crawling " + url)
  print("Documents Seen Until This URL: " + str(len(frontierQueue)))

  br.open(url)

  for link in br.links():

    #Validate URL
    if (validators.url(link.url)):
      #print("Generated: " + link.url)
      frontierQueue = frontierQueueInsert(frontierQueue, link.url)

    #If not valid try to generate with the relative path
    else:
      nUrl = generateRelativePath(link)
      if (nUrl != None):
        frontierQueue = frontierQueueInsert(frontierQueue, nUrl)
  
  print("\n")
  return frontierQueue
```

```
lastCrawledIndex = 0

while (len(frontierQueue) < maxDocuments):
  crawl_url(br, frontierQueue[lastCrawledIndex], frontierQueue)
  lastCrawledIndex = lastCrawledIndex + 1

if (len(set(frontierQueue)) == len(frontierQueue) and len(frontierQueue) == maxDocuments):
  print("Finished with "+str(maxDocuments)+" unique documents.")

mechanizeDocuments = frontierQueue
```

# **Crawling from scratch**

```
import requests
import re
from urllib.parse import urlparse
import time
import urllib.robotparser
```

```
class ScratchCrawler():
  def __init__(self, originURL, max_docs):
    self.origin_url = originURL;
    self.visited = []
    self.max_docs = max_docs
    self.disallowed_robots = []
    
  def get_html(self, url):

    print("Getting: " + url)
    print("Documents Seen Until This URL: " + str(len(self.visited)))
    try:
      dom = requests.get(url, timeout = 5)

      content_type = dom.headers['Content-Type']
      print("Successful request. Content type: " + str(content_type))
    except Timeout:
      print('The request timed out.')
      return "ERR"
    except Exception as e:
      print(e)
      return "ERR"
    return dom.content.decode('latin-1')

  def get_links(self, url):
    html_code = self.get_html(url)

    if html_code == "ERR":
      return []
    else:   
      parsed_html = urlparse(url)
      base = f"{parsed_html.scheme}://{parsed_html.netloc}"  
              
      links = re.findall('''<a\s+(?:[^>]*?\s+)?href="([^"]*)"''', html_code)

      for i, link in enumerate(links):
        if not urlparse(link).netloc:
          n_link = base + link
          links[i] = n_link

      return links

  def crawl(self, url):
    # If an error or timeout is produced, result_links will be empty
    # and nothing will happen

    result_links = self.get_links(url)
    for link in result_links:
      if link not in self.visited and len(self.visited) < self.max_docs:
        self.visited.append(link)
    
  
  def start(self):
    rp = urllib.robotparser.RobotFileParser()
    parsed_html = urlparse(self.origin_url)
    base = f"{parsed_html.scheme}://{parsed_html.netloc}"
    rp.set_url(base+"/robots.txt")
    rp.read()
    rrate = rp.request_rate('python-requests/2.21.0')
    
    self.visited.append(self.origin_url)

    n_crawled = 0
    while len(self.visited) < self.max_docs:
      # Make sure we are allowed to fetch that link (etiquette)
      if rp.can_fetch('python-requests/2.21.0',self.visited[n_crawled]):
        self.crawl(self.visited[n_crawled])
      n_crawled = n_crawled + 1

      if not len(self.visited) == self.max_docs:
        #Sleep until next request (politeness & etiquette)
        print("\n")
        print("Waiting " + str(rrate.seconds if rrate else 5) + " seconds until next request...")
        print("\n")
        time.sleep(rrate.seconds if rrate else 5)

      
```

```
originURL = "https://store.steampowered.com/games/"
max_docs = 500
s_crawler = ScratchCrawler(originURL, max_docs)
s_crawler.start()
print("\n")
s_crawler.visited
```

```
docs = s_crawler.visited
scratch_docs = []

for url in docs:
  if '/app/' in url:
    scratch_docs.append(url)
scratch_docs
```

```
class SteamGame():
    def __init__(self, game=None, category=None, price=None, developer=None, reviews=None, release_date=None, image_url=None, url=None):
      self.game = clean_string(game)
      self.category = clean_string(category)
      self.price = clean_string(price)
      self.developer = clean_string(developer)
      self.reviews = clean_string(reviews)
      self.release_date = clean_string(release_date)
      self.image_url = clean_string(image_url)
      self.url = url

    def __str__(self):
      return str(self.game) + ";" + str(self.category) + ";" + str(self.price) + ";" + str(self.developer) + ";" + str(self.reviews) + ";" + str(self.release_date) + ";" + str(self.image_url) + ";" + str(self.url)



def dict2SteamGame(element):
  return SteamGame(element.get('game') if element.get('game') else None,
                   element.get('category') if element.get('category') else None,
                   element.get('price') if element.get('price') else None,
                   element.get('developer') if element.get('developer') else None,
                   element.get('reviews') if element.get('reviews') else None,
                   element.get('release_date') if element.get('release_date') else None,
                   element.get('image_url') if element.get('image_url') else None,
                   element.get('url') if element.get('url') else None
                   )

def clean_string(myString):

  if (isinstance(myString, list)):
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in myString:  
        str1 += ele   
    
    # return string   
    myString = str1  

  if myString:
    myString = ' '.join(myString.split())

    # Si devuelve HTML (contiene "<") es porque no ha encontrado lo que queriamos, 
    # asi que directamente nos lo cargamos
    if "<" in myString:
      myString = None
    else:
      # Sustituir el &amp por un ampersant, que si no fastidia el csv
      if "&amp;" in myString:
        myString = myString.replace('&amp;','&')

  return myString
```
