```
!pip install mechanize
!pip install validators
import mechanize
import validators
from urllib.parse import urlparse
```

```
maxDocuments = 500
originURL = "https://es.wikipedia.org/wiki/Wikipedia:Portada"
frontierQueue = [originURL]

br = mechanize.Browser()
```

```
br.open(originURL)
br.title()
```

```
def generateRelativePath(link):
  newURL = "http://" + urlparse(link.base_url).netloc + link.url
  if (validators.url(newURL) == True and link.url[0] == '/'):
    return newURL
```

```
def frontierQueueInsert(frontierQueue, url):
  toInsert = True

  if (url in frontierQueue):
    toInsert = False
  if (len(frontierQueue) >= maxDocuments):
    toInsert = False

  if (toInsert == True):
    frontierQueue.append(url)
  return frontierQueue
```

```
def crawl_url(br, url, frontierQueue):
  print("Crawling " + url)
  print("Documents Seen Until This URL: " + str(len(frontierQueue)))

  br.open(url)

  for link in br.links():

    #Validate URL
    if (validators.url(link.url)):
      #print("Generated: " + link.url)
      frontierQueue = frontierQueueInsert(frontierQueue, link.url)

    #If not valid try to generate with the relative path
    else:
      nUrl = generateRelativePath(link)
      if (nUrl != None):
        frontierQueue = frontierQueueInsert(frontierQueue, nUrl)
  
  print("\n")
  return frontierQueue
```

```
lastCrawledIndex = 0

while (len(frontierQueue) < maxDocuments):
  crawl_url(br, frontierQueue[lastCrawledIndex], frontierQueue)
  lastCrawledIndex = lastCrawledIndex + 1

if (len(set(frontierQueue)) == len(frontierQueue) and len(frontierQueue) == maxDocuments):
  print("Finished with "+str(maxDocuments)+" unique documents.")
```
