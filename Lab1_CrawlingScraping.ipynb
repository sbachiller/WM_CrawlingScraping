```
maxDocuments = 500
originURL = "https://www.uam.es/"
frontierQueue = []

br = mechanize.Browser()
```

```
br.open(originURL)
br.title()
```

```
def generateRelativePath(link):
  newURL = "http://" + urlparse(link.base_url).netloc + link.url
  if (validators.url(newURL) == True and link.url[0] == '/'):
    return newURL
```

```
for link in br.links():
  print("Found: " + link.url)

  #Validate URL
  if (validators.url(link.url)):
    print("Generated: " + link.url)

  #If not valid try to generate with the relative path
  else:
    print("Generated: " + str(generateRelativePath(link)))

  print("\n")

frontierQueue
```
